<b>Правила именования переменных:</b>

- Имена переменных не должны повторять уже встроенные в Python имена, например, print, str, in и т.д.

### Операторы сравнения
Возвращают <b>True/False</b> ка результат сравнения

Сравнение высказываний. Логическое высказывание (предикат) - любое повествовательное предложение, в отношении которого можно однозначно сказать, истинно оно или ложно. 
- Высказывание: «6 — четное число». Истинно или ложно? Очевидно, что истинно. А высказывание: «6 больше 19»? Хм. Высказывание ложно. Никакого подвоха в этом нет.
- Является ли высказыванием фраза: «у него голубые глаза»? Хочется спросить, у кого? Однозначности в этой фразе нет, поэтому она не является высказыванием.
- Высказывания можно комбинировать. Высказывания «Петров – врач», «Петров – шахматист» можно объединять с помощью связок И, ИЛИ

```python
#> greater
#< smaller
#>= greater&equal
#<= smaller&equal
#== equal
#!= not equal

print (7<10)
print (7<0)
print (7==7)
print (7!=7)
#type(True) #will return 'bool' - its another type of data (boolean or logical)
#True interpreted in Py as ==1, False as ==0

print (0.1 + 0.1 == 0.2) #True
print (0.1 + 0.1 + 0.1 == 0.3) #False! There are some 'out of sense' peculiarities of comparison
#(see IEEE 754 standard: http://www.softelectro.ru/ieee754.html)
```

- Для Python истинным или ложным может быть не только логическое высказывание, но и объект. В Python любое число, не равное нулю, или непустой объект интерпретируется как истина.
- Числа, равные нулю, пустые объекты и специальный объект None (type = NoneType) интерпретируются как ложь.
- У Python есть три логических оператора and, or, not.

### Особенности сравнения и выполнения логических операций

- Для вычисления оператора and Python вычисляет операнды слева направо и возвращает первый объект, имеющий ложное значение. Посмотрим на столбец and таблицы истинности. Какая закономерность? Если среди операндов (X,Y) есть ложный, то получим ложное значение, но вместо ложного значения для операндов-объектов Python возвращает первый ложный операнд, встретившийся в выражении, и дальше вычисления НЕ производит. Это называется вычислением по короткой схеме ("ленивое AND"). Если Python не удается найти ложный объект-операнд, то он возвращает крайний правый операнд.
- Логический оператор or действует похожим образом, но для объектов-операндов Python возвращает первый объект,имеющий истинное значение. Python прекратит дальнейшие вычисления, как только будет найден первый объект, имеющий истинное значение.
- Отдельного <b><a href='http://www.softelectro.ru/ieee754.html' a>разговора</b> заслуживает сравнение вещественных чисел, т.к. оно может привести к неожиданным, на первый взгляд, результатам (см. стандарт IEEE 754).

### Многомерные списки

<b><a href='https://foxford.ru/wiki/informatika/mnogomernye-spiski-v-python' a>Источник.</b> Часто приходится хранить прямоугольные таблицы с данными (матрицаы или двумерные массивы). В Python таблицу можно представить в виде списка строк, каждый элемент которого является в свою очередь списком, например, чисел. Например, создать числовую таблицу из двух строк и трех столбцов можно так:

```python
A = [ [1, 2, 3], [4, 5, 6] ]

# Здесь первая строка списка A[0 ] является списком из чисел [1, 2, 3]. То есть:
A[0][0] == 1, A[0][1] == 2, A[0][2] == 3
A[1][0] == 4, A[1][1] == 5, A[1][2] == 6
```

Для обработки и вывода списка обычно используется два вложенных цикла. Первый цикл по номеру строки, второй цикл по элементам внутри строки. Вывод двумерного числового списка построчно, разделяя числа пробелами внутри одной строки:

```python
for i in range(len(A)):
    for j in range(len(A[i]):
       print(A[i][j], end = ' ')
    print()
                   
# То же самое, но циклы не по индексу, а по значениям списка:
for row in A:
    for elem in row:
       print(elem, end = ' ')
    print()
#  Для вывода одной строки - метод join:
for row in A:
    print(' '.join(map(str, row)))
                   
# Два вложенных цикла для подсчета суммы всех чисел в списке:
S = 0
for i in range(len(A)):
    for j in range(len(A[i])):
        S += A[i][j]
                   
# Или то же самое с циклом не по индексу, а по значениям строк:
S = 0
for row in A:
    for elem in row:
        S += elem
```

СОЗДАНИЕ ДВУМЕРНОГО СПИСКА. Пусть даны два числа: количество строк n и количество столбцов m. Необходимо создать список размером n×m, заполненный нулями. Очевидное решение оказывается неверным:

```python
A = [[0] * m] * n
```

В этом легко убедиться, если присвоить элементу A[0][0] значение 1, а потом вывести значение другого элемента A[1][0] — оно тоже будет равно 1! Дело в том, что [0] * m возвращает ccылку на список из m нулей. Но последующее повторение этого элемента создает список из n элементов, которые являются ссылкой на один и тот же список (точно так же, как выполнение операции B = A для списков не создает новый список), поэтому все строки результирующего списка - одна и та же строка.

Т.е. двумерный список нельзя создавать операцией повторения одной строки. Сначала создадим список из n элементов (из n нулей). Затем сделаем каждый элемент списка ссылкой на другой одномерный список из m элементов:

```python
A = [0] * n
for i in range(n):
    A[i] = [0] * m
    
# Другой способ: создать пустой список, потом n раз добавить в него новый элемент, являющийся списком-строкой:
A = []
for i in range(n):
    A.append([0] * m)
    
# Проще работать с генератором: создать список из n элементов, каждый из которых будет списком, состоящих из m нулей:
A = [[0] * m for i in range(n)]

'''Тут каждый элемент создается независимо от остальных (заново конструируется список [0] * m для заполнения 
очередного элемента списка), а не копируются ссылки на один и тот же список.'''
```

Варианты решения. Элементы, выше диагонали – это элементы, для которых i<j, а для элементов ниже диагонали i>j. Т.е. можно сравнивать значения i и j и по ним определять значение A[i][j]. Получаем следующий алгоритм:

```python
for i in range(n):
    for j in range(n):
        if i < j:
            A[i][j] = 0
        elif i > j:
            A[i][j] = 2
        else:
            A[i][j] = 1
            
'''Данный алгоритм плох, поскольку выполняет одну или две инструкции if для обработки каждого элемента. 
Тут лучше обойтись вообще без условных инструкций. Сначала заполним диагональ, для чего понадобится один цикл:'''
for i in range(n):
    A[i][i] = 1
    
'''Затем заполним значением 0 все элементы выше диагонали, для чего нам понадобится в каждой из строк 
с номером i присвоить значение элементам A[i][j] для j=i+1, …, n-1. Здесь нам понадобятся вложенные циклы:'''
for i in range(n):
    for j in range(i + 1, n):
        A[i][j] = 0

# Аналогично присваиваем значение 2 элементам A[i][j] для j=0, …, i-1:
for i in range(n):
    for j in range(0, i):
        A[i][j] = 2
        
# Можно также внешние циклы объединить в один и получить еще одно, более компактное решение:
for i in range(n):
    for j in range(0, i):
        A[i][j] = 2
    A[i][i] = 1
    for j in range(i + 1, n):
        A[i][j] = 0 
        
'''Решение, использующее операцию повторения списков для построения очередной строки списка. 
i-я строка списка состоит из i чисел 2, затем идет одно число 1, затем идет n-i-1 число 0:'''
for i in range(n):
    A[i] = [2] * i + [1] + [0] * (n - i - 1)
    
# Можно заменить цикл на генератор:
A = [[2] * i + [1] + [0] * (n - i - 1) for i in range(n)]

# Generator can directly accept file as iterable object and perform some modifications with elements:
line = [line.strip ().split ('\t') for line in open ('my_file.csv', 'r')]

# Easy way to pass each i-th element of list_1 to list_2:
list_2 = list_1[::i]
```

### Тип данных словарь (dict)

<b>Словарь</b> (dict) – неупорядоченная изменяемая коллекция или «список» с произвольными ключами неизменяемого типа. Как ис список, словарь может хранить различные типы, в т.ч. вложенные структуры данных. В большинстве случаев поиск по ключам словаря не зависит от времени (поиск по словарю из 100 сотрудников будет занимать почти такое же время, как и по словарю из 100 тысяч сотрудников) - т.е. алгоритмическая сложность по времени = <b>О(1)</b>. И это гораздо быстрее, чем перебор элементов аналогичного списка - линейная сложность <b>O(n)</b>

Ключами словаря могут быть значения только hashable типов, то есть типов, у которых может быть получен хэш (для этого у них должен быть метод __hash__()). Обычно, <b>hash ()</b> есть только у неизменяемых типов. Поэтому такие типы, как список (list), множество (set) и сам словарь (dict), не могут быть ключами словаря. А вот кортеж (tuple), строка (string) или число (int, float) - могут. Чаще всего это строки.

Часто словари используются, если требуется найти частоту встречаемости элементов в последовательности (списке, строке, кортеже). Функция, которая возвращает словарь со статистикой встречаемости элементов в последовательности:

```python
def histogram(s):
    d = dict()
    for c in s:
        if c not in d:
            d[c]=1
        else:
            d[c]+= 1
    return d

# Результат вызова функции histogram() для списка, строки, кортежа соответственно:
histogram([2,5,6,5,4,4,4,4,3,2,2,2,2])
{2: 5, 3: 1, 4: 4, 5: 2, 6: 1}

histogram("ywte3475eryt3478e477477474")
#{'4': 6, '8': 1, 'e': 3, '3': 2, '7': 7, '5': 1, 'r': 1, 'y': 2, 'w':1, 't': 2}

histogram((5,5,5,6,5,'r',5))
#{5: 5, 6: 1, 'r': 1}

# Создание и сортировка словаря по значению:
d={"t1":2, "t6":5, "t9":1}
d
{'t9': 1, 't6': 5, 't1': 2}
sorted(d)
['t1', 't6', 't9']
d.get('t1')
2
sorted(d, key=d.get)
['t9', 't1', 't6']
```

```python
my_dict = {i: i * i for i in range(10)} 
print (my_dict)
```
